```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Loading of the Inbound dataset
inbound = pd.read_csv(r"C:\Users\frifi\OneDrive\Desktop\School Mat UHasselt\Third Semester\Project Data Science\Data sets\Inbound\Original_inbound_final_1.csv")
inbound
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12, 8))

# Stacked bar plot for 'Room_Temp pallets' and 'Frigo_Temp pallets'
inbound[['Room_Temp pallets', 'Frigo_Temp pallets']].plot(kind='bar', stacked=True, width=0.8, colormap='viridis')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Number of Pallets')
plt.title('Composition of Room_Temp and Frigo_Temp Pallets over Time')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'inbound'
plt.figure(figsize=(12, 8))

# Grouped bar plot for 'Room_Temp pallets' and 'Frigo_Temp pallets'
sns.barplot(x='Date', y='Room_Temp pallets', data=inbound, color='blue', label='Room Temp Pallets')
sns.barplot(x='Date', y='Frigo_Temp pallets', data=inbound, color='orange', label='Frigo Temp Pallets')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Number of Pallets')
plt.title('Room_Temp and Frigo_Temp Pallets over Time')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'inbound'
plt.figure(figsize=(12, 8))

# Grouped bar plot for 'Room_Temp pallets' and 'Frigo_Temp pallets'
sns.barplot(x='Date', y='Room_Temp pallets', data=inbound, color='blue', label='Room Temp Pallets')
sns.barplot(x='Date', y='Frigo_Temp pallets', data=inbound, color='orange', label='Frigo Temp Pallets')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Number of Pallets')
plt.title('Room_Temp and Frigo_Temp Pallets over Time')

# Adding legend
plt.legend()

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python

plt.figure(figsize=(10, 6))

# Scatter plot
sns.scatterplot(x='Total number of pallets', y='Actual number of workers', data=inbound,  palette='viridis')

# Adding labels and title
plt.xlabel('Total number of pallets')
plt.ylabel('Actual number of workers')
plt.title('Scatter Plot of Number of Pallets vs. Actual number of Workers')

# Display the plot
plt.show()
```

```python
#Plot of Pallets and workload
plt.figure(figsize=(10, 6))

# Scatter plot
sns.scatterplot(x='Total number of pallets', y='Workload(hrs)', data=inbound, palette='viridis')

# Adding labels and title
plt.xlabel('Total number of pallets')
plt.ylabel('Workload(hrs)')
plt.title('Scatter Plot of Pallets vs. Workload')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting histograms for numerical variables
plt.figure(figsize=(10, 8))
plt.hist([inbound['Room_Temp pallets'], inbound['Frigo_Temp pallets']], bins=20, alpha=0.7, label=['Room Temp', 'Frigo Temp'])
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Distribution of Room and Frigo Temperatures')
plt.legend(loc='upper right')
plt.show()
```

```python

# Plotting a bar plot for the distribution of Truck frequency
plt.figure(figsize=(8, 6))
sns.countplot(x='Truck Frequency', data=inbound)
plt.title('Distribution of Truck Frequency')
plt.show()
```

```python

# Plotting a bar plot for the distribution of Trucks 
plt.figure(figsize=(8, 6))
sns.countplot(x='Total_trucks', data=inbound)
plt.title('Distribution of Trucks')
plt.show()
```

```python
# Plotting a scatter plot for the relationship between Room_Temp and Frigo_Temp pallets
plt.figure(figsize=(8, 6))
plt.scatter(inbound['Room_Temp pallets'], inbound['Frigo_Temp pallets'])
plt.xlabel('Room Temp')
plt.ylabel('Frigo Temp')
plt.title('Relationship between Room and Frigo Temperatures')
plt.show()
```

```python
# Plotting a scatter plot for the relationship between Pallets and Workload
plt.figure(figsize=(8, 6))
plt.scatter(inbound['Total number of pallets'], inbound['Workload(hrs)'])
plt.xlabel('Total number of pallets')
plt.ylabel('Workload(hrs)')
plt.title('Relationship between Total number of pallets and Workload(hrs)')
plt.show()
```
